```python
rate = pd.DataFrame(marketing_customer_analysis['Response'].value_counts()).reset_index()
rate
```

```python
rate_df = pd.DataFrame(marketing_customer_analysis['Response'].value_counts()/len(marketing_customer_analysis)*100).reset_index()
rate_df
```

```python
fig, ax = plt.subplots()
ax.bar(x = rate['index'], height= rate['Response'])
ax.set_xlabel("Responses")
ax.set_ylabel("Count")
ax.set_title("total number of responses")
plt.show()
```

```python
answer_rate = marketing_customer_analysis.copy()
answer_rate = pd.DataFrame(marketing_customer_analysis[['Response','Sales Channel',]].value_counts(dropna=False).reset_index())
answer_rate.columns = ['Response','Sales Channel', 'Rate']
answer_rate 
```

```python

answer_rate['Rate'] = (answer_rate['Rate']/marketing_customer_analysis.shape[0])*100
answer_rate
```

```python
fig, ax = plt.subplots()
ax.set_ylabel("Count")
ax.set_title("Rate of Responses by Channel %")
ax.bar(x = answer_rate['Sales Channel'], height = answer_rate['Rate'], color=['steelblue', 'grey', 'green', 'red'])

plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(x = answer_rate['Response'], height= answer_rate['Rate'])
ax.bar(x = answer_rate['Response'], height= answer_rate['Rate'])
ax.set_xlabel("Responses")
ax.set_ylabel("Count")
ax.set_title("total number of responses")
plt.show()
```

```python
answer_rate2 = marketing_customer_analysis.copy()
answer_rate2 = pd.DataFrame(marketing_customer_analysis[['Response','Total Claim Amount',]].value_counts(dropna=False).reset_index())
answer_rate2.columns = ['Response','Total Claim Amount', 'Rate']
answer_rate2 
```

```python
fig, ax = plt.subplots(figsize=(10,5))
# sns.barplot(x=marketing_customer_analysis['index'].value_counts().index, height = rate_df['Response'].values, hue='Sales Channel')
sns.barplot(x='Rate', y='Total Claim Amount', data=answer_rate2 )
ax.set_xlabel("Responses")
ax.set_ylabel("Count")
ax.set_title("% COUNT of responses by total claim amount")
plt.show()
```

```python
marketing_customer_analysis['Income'].value_counts()
```

```python
answer_rate1 = marketing_customer_analysis.copy()
answer_rate1 = pd.DataFrame(marketing_customer_analysis[['Response','Income',]].value_counts(dropna=False).reset_index())
answer_rate1.columns = ['Response','Income', 'Rate']
answer_rate1 
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.set_xlabel("Income")
ax.set_ylabel("Count")
ax.set_title("% Average of responses by INCOME")
# sns.barplot(x=marketing_customer_analysis['index'].value_counts().index, height = rate_df['Response'].values, hue='Sales Channel')
sns.boxplot(x='Response', y='Income', data=answer_rate1)
plt.show()

plt.show()
```

```python
fix, ax = plt.subplots( figsize=(10,5))
sns.boxplot(data=answer_rate1, x= 'Income', ax=ax)
plt.tight_layout()
plt.show()
```

```python
fix, ax = plt.subplots( figsize=(10,5))
sns.scatterplot(data=answer_rate1, x= 'Income', y='Rate', hue='Response', ax=ax)
plt.tight_layout()
plt.show()
```
